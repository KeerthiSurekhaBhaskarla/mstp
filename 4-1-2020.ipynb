# 3

```python
import matplotlib.pyplot as plt
x=df['sepal_length']
y=df['petal_length']
plt.title('scatter plot')
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.scatter(x,y)
```

# 4

```python
x=df['sepal_length']
y=df['petal_length']
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
plt.barh(x,y)
plt.show()
```
